```python
xb_x1 = xb1.iloc[:,[1,2,3,4,5,8,11,14,27,28]]
xb_x1.describe(include='all')
```

```python
xb_y1 = xb1.iloc[:,7]
xb_y1.describe()
```

```python
                                                    ####### Sampling #######
import sklearn
from sklearn.model_selection import train_test_split

# cr_x_train,cr_x_test,cr_y_train,cr_y_test = train_test_split(cr_x,cr_y,train_size = .8,test_size = .2,random_state = 256) 
xb_x1_train,xb_x1_test,xb_y1_train,xb_y1_test = train_test_split(xb_x1,xb_y1,test_size = .20,random_state = 101)   
## (random_state = set.seed(121) We Can Change As We Wont 101 To 123,145,256,.......n etc)

## X_Train_Test ##
print(xb_x1_train.shape)       ## Shows The Total 80% Number of Rows & Columns For Credit Risk Data For X Variables
print(xb_x1_test.shape)        ## Shows The Total 20% Number of Rows & Columns For Credit Risk Data For X Variables

## Y_Train_Test ##
print(xb_y1_train.shape)       ## Shows The Total 80% Number of Rows & Columns For Credit Risk Data For Y Variables
print(xb_y1_test.shape)        ## Shows The Total 20% Number of Rows & Columns For Credit Risk Data For Y Variables
```

```python
                                            ##### Decision Tree Model Bulding #####

from sklearn.tree import DecisionTreeClassifier    ## Importing Linear Regression
dtree = DecisionTreeClassifier(min_samples_split=100)

dtree.fit(xb_x1_train,xb_y1_train)
pred_xb1 = dtree.predict(xb_x1_test)

from sklearn.metrics import confusion_matrix
print(confusion_matrix(pred_xb1,xb_y1_test))

tab1 = confusion_matrix(pred_xb1,xb_y1_test)
accurary = sum(tab1.diagonal())/tab1.sum()               ### Get The Accuracy
print("Accuracy Of The Model is :: ",accurary)

print(dtree.feature_importances_)              ## Shows the Important Column Attribute For The Model By Defing Value
print(xb_x1_train.head())
```

```python
                                               ##### Best Module #####                                           
                                       ##### Random Forest Model Bulding #####

from sklearn.ensemble import RandomForestClassifier        ## Importing Linear Regression
dtree = RandomForestClassifier(min_samples_split=100)

dtree.fit(xb_x1_train,xb_y1_train)
pred_xb1 = dtree.predict(xb_x1_test)

from sklearn.metrics import confusion_matrix
print(confusion_matrix(pred_xb1,xb_y1_test))

tab1 = confusion_matrix(pred_xb1,xb_y1_test)
accurary = sum(tab1.diagonal())/tab1.sum()               ### Get The Accuracy
print("Accuracy Of The Model is :: ",accurary)

print(dtree.feature_importances_)              ## Shows the Important Column Attribute For The Model By Defing Value
print(xb_x1_train.head())
```
